La gráfica anterior incluye nombres de los ejes (etiquetas) que aparecerán en la gráfica, al igual que el título del gráfico.

```python
# @title Gráfica x vs y
x = [-2, -1, 0, 1, 2] #Valores en x, donde comienzan y terminan
y = [3, 4, -1, 2, -6] #Valores en y, tienen que haber tantos como en x
plt.plot(x, y, 'r^') # o, -, --, s(cuadros), v(viñeta), .(puntos), o(puntos), ^(triángulo)
plt.xlabel('Tiempo(seg)')
plt.ylabel('Amplitud(voltios)')
plt.title('Gráfica de prueba')
plt.grid() #Cuadricula
plt.show()
```

```python
# @title Gráfica x vs y
import numpy as np
plt.figure(figsize=(10, 5))

plt.subplot(2, 2, 1) #Filas, columnas, posición
t = np.arange(-2, 3, 0.1)
z = 3*t**2 - 5*t + 1
plt.plot(t, z, 'k.')
plt.grid()

plt.subplot(2, 2, 2)
z2 = np.sin(t)
plt.plot(t, z2, 'r.')

plt.subplot(2, 2, 3)
z3 = np.tan(t)
plt.plot(t, z3, 'g.')

plt.subplot(2, 2, 4)
x = [-2, -1, 0, 1, 2]
y = [3, 5, -1, 2, -6]
plt.plot(x, y, 'ro')
plt.grid()

plt.show()
```

```python
# @title Gráfica x vs y
import numpy as np
plt.figure(figsize=(10, 5))

plt.subplot(2, 2, 1) #Filas, columnas, posición
t = np.arange(-2, 3, 0.1)
z = 3*t**2 - 5*t + 1
plt.plot(t, z, 'k.')
plt.grid()

plt.subplot(2, 2, (2, 4))
z2 = np.sin(t)
plt.plot(t, z2, 'r.')

plt.subplot(2, 2, 3)
z3 = np.tan(t)
plt.plot(t, z3, 'g.')

plt.show()
```

```python
# @title Gráfica x vs y
import numpy as np
plt.figure(figsize=(10, 5))
plt.subplots_adjust(wspace=0.4, hspace=0.4) # Ajusta el espacio horizontal y vertical

plt.subplot(2, 2, 1)
plt.title('Gráfica 1')
t = np.arange(-2, 3, 0.1)
z = 3*t**2 - 5*t + 1
plt.plot(t, z, 'k.')
plt.grid()

plt.subplot(2, 2, 2)
plt.title('Gráfica 2')
z2 = np.sin(t)
plt.plot(t, z2, 'r.')

plt.subplot(2, 2, (3, 4))
plt.title('Gráfica 3')
z3 = np.tan(t)
plt.plot(t, z3, 'g.')

plt.show()
```

# ***Gráficos de barras***

```python
import matplotlib.pyplot as plt
import numpy as np
```

```python
dias = ['Lunes', 'Martes', 'Miercoles', 'Jueves', 'Viernes']
asistencia = [29, 15, 22, 24, 25]

plt.bar(dias, asistencia, color='#A003FB', width=0.8)
plt.show()
```

```python
dias = ['Lunes', 'Martes', 'Miercoles', 'Jueves', 'Viernes']
asistencia = [29, 15, 22, 24, 25]

plt.barh(dias, asistencia, color='#A003FB')
# Length
promedio = sum(asistencia) / len(asistencia)
plt.axvline(promedio, color='red', linestyle='--', label=f'Promedio: {promedio:.2f}')

plt.xlabel('Asistencia')
plt.ylabel('Días')
plt.title('Asistencia por día')
plt.legend()
plt.show()
```

```python
dias = ['Lunes', 'Martes', 'Miercoles', 'Jueves', 'Viernes']
asistencia = [29, 15, 22, 24, 25]
promedio = asistencia.mean()

plt.bar(dias, asistencia, color='#A003FB', width=0.8)
plt.show()
```
