```python
# Install dependencies as needed:
# pip install kagglehub[pandas-datasets]
import kagglehub
from kagglehub import KaggleDatasetAdapter

# Set the path to the file you'd like to load
file_path = "Data_completa.csv"

# Load the latest version
df = kagglehub.load_dataset(
  KaggleDatasetAdapter.PANDAS,
  "williamrrubio/data-siniestros-bogot-2015-2021",
  file_path,
)
```

```python
j = 1
plt.figure(figsize=(15, 8))
plt.subplots_adjust(wspace=0.4, hspace=0.4) # Ajusta el espacio horizontal y vertical
for i in df.columns:
    if i == 'choque' or i == 'clasevehiculo' or i == 'servicio' or i == 'causa':
        continue
    else:
        plt.subplot(3, 2, j)
        plt.hist(df[i], orientation = 'horizontal')
        plt.title(i)
    j += 1

plt.show()
```
